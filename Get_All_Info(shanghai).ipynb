```python
# Import 
from splinter import Browser
from bs4 import BeautifulSoup
import pandas as pd
from webdriver_manager.chrome import ChromeDriverManager
import requests
```

```python
# Set the executable path and initialize Splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Import Links to array 
link_df = pd.read_csv('links.csv')

# Pandas DataFrame to Array 
link_array = link_df.to_numpy()

link_df
```

```python

all_info = []
for z in range(2580,2610):
    
    URL = link_array[z][0]
    r = requests.get(URL, timeout=None)
    
    browser.visit(URL)
    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')
    content = soup.find('div', class_='content')
    source_info = []
    detail = content.find('ul')
    for li in detail.find_all('li'):
        info = [li.text]
        source_info.append(info)
    deal_type = soup.find('div', class_='transaction')
    detail1 = deal_type.find('ul')
    for li1 in detail1.find_all('li'):
        info1 = [li1.text]
        source_info.append(info1)
    source_info1=[]
    views = soup.find('div', class_='msg')
    for li2 in views.find_all('span'):
        info2=[li2.text]
        source_info1.append(info2)
    df1 = pd.DataFrame(source_info)
    final_info_draft = []
    for i in range(0,len(source_info)):
        final_info_draft.append(df1[0][i][4:])
    totalprice = soup.find('div', class_='price').span.text
    unitprice = soup.find('div', class_='price').b.text
    dealdate = soup.find('div', class_='wrapper').span.text
    title = soup.find('div', class_='wrapper').h1.text
    final_info_draft.append(totalprice)
    final_info_draft.append(unitprice)
    final_info_draft.append(dealdate)
    final_info_draft.append(title)
    final_info_draft.append(source_info1[0][0])
    final_info_draft.append(source_info1[1][0])
    final_info_draft.append(source_info1[2][0])
    final_info_draft.append(source_info1[3][0])
    final_info_draft.append(source_info1[4][0])
    final_info_draft.append(source_info1[5][0])
    all_info.append(final_info_draft)
    print(z)
    
final_df = pd.DataFrame(all_info)   
```
